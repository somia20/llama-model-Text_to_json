**SPLITTING DATA INTO TRAIN AND VALIDATION**

```python
import pandas as pd
from sklearn.model_selection import train_test_split

def process_and_split_csv(csv_path):
    # Load the CSV data into a DataFrame
    df = pd.read_csv(csv_path)

    # Split the DataFrame into training (80%) and validation (20%) sets
    train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)

    # Save the split data into separate CSV files
    train_path = csv_path.replace('.csv', '_train.csv')
    val_path = csv_path.replace('.csv', '_val.csv')

    train_df.to_csv(train_path, index=False)
    val_df.to_csv(val_path, index=False)

    print(f"Training data saved to: {train_path}")
    print(f"Validation data saved to: {val_path}")

# Example usage
csv_path = '/content/text_json.csv'  # Replace with the actual CSV file path
process_and_split_csv(csv_path)
```

**CONVERTING DATA IN ALPACCA FORMAT AND TRAINING**

```python
# format the text columns using the Alpaca format
alpaca_prompt = """Below is an instruction that describes a task, paired with an input that provides further context. Write a response that appropriately completes the request.\n\n

### Instruction:\n
{}\n\n

### Input:\n
{}\n
{}\n\n

### Response:\n
{}"""

EOS_TOKEN = tokenizer.eos_token
def formatting_prompts_func(examples):
    instructions = examples["instruction"]
    inputs = examples["context"]
    outputs      = examples["response"]
    texts = []
    for instruction, input, output in zip(instructions, inputs, outputs):
        text = alpaca_prompt.format(instruction, input, EOS_TOKEN, output)
        texts.append(text)
    return { "text" : texts, }

# map the formatting function on train and eval sets
train_dataset = train_dataset.map(formatting_prompts_func, batched = True,)
eval_dataset = eval_dataset.map(formatting_prompts_func, batched = True,)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 32,                                 # Choose any number > 0 ! Suggested 8, 16, 32, 64, 128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj",],
    lora_alpha = 16,
    lora_dropout = 0,                       # Supports any, but = 0 is optimized
    bias = "none",                          # Supports any, but = "none" is optimized
    use_gradient_checkpointing = "unsloth", # True or "unsloth" for very long context
    random_state = 3407,
    use_rslora = True,                      # We support rank stabilized LoRA
    loftq_config = None,                    # And LoftQ
)
```

```python
from transformers import TrainerCallback

class CustomCallback(TrainerCallback):
    def __init__(self):
        self.train_losses = []
        self.eval_losses = []

    def on_log(self, args, state, control, logs=None, **kwargs):
        if logs is not None:
            if 'loss' in logs:
                self.train_losses.append(logs['loss'])
            if 'eval_loss' in logs:
                self.eval_losses.append(logs['eval_loss'])

callback = CustomCallback()
```

```python
import os
os.environ['WANDB_MODE'] = 'disabled'
```

```python
from transformers import TrainingArguments, Trainer
from unsloth import is_bfloat16_supported

training_args = TrainingArguments(
    per_device_train_batch_size=2,        # how many samples to train at a time per step per device
    gradient_accumulation_steps=4,        # total steps will be divided by this
    warmup_steps=2,                       # steps to go from 0 to specified lr
    num_train_epochs=4,                   # no of samples per step * this would be total steps
    learning_rate=2e-4,
    fp16=not is_bfloat16_supported(),
    bf16=is_bfloat16_supported(),
    logging_steps=1,
    optim="adamw_8bit",
    weight_decay=0.01,
    lr_scheduler_type="linear",
    seed=3407,
    output_dir="outputs",
    evaluation_strategy="steps",          # Evaluate during training
    eval_steps=1,                         # Evaluate every 1 step
    save_strategy="steps",
    save_steps=10,                        # save model state every 10 steps
    load_best_model_at_end=True,          # Load the best model when finished training
)


trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=train_dataset,          # pass in the training split
    eval_dataset=eval_dataset,            # pass in the eval split
    args=training_args,
    dataset_text_field = "text",
    callbacks=[callback]                  # our custom callback
)

# check GPU memory before training
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")

# start training
trainer_stats = trainer.train()

# check GPU memory after training
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory         /max_memory*100, 3)
lora_percentage = round(used_memory_for_lora/max_memory*100, 3)
print(f"{trainer_stats.metrics['train_runtime']} seconds used for training.")
print(f"{round(trainer_stats.metrics['train_runtime']/60, 2)} minutes used for training.")
print(f"Peak reserved memory = {used_memory} GB.")
print(f"Peak reserved memory for training = {used_memory_for_lora} GB.")
print(f"Peak reserved memory % of max memory = {used_percentage} %.")
print(f"Peak reserved memory for training % of max memory = {lora_percentage} %.")
```

**DATASET CREATION**

```python
# Install required packages
!pip install groq

from groq import Groq
import pandas as pd
import json
import time

# Set your Groq API key (replace with your actual key)
GROQ_API_KEY = ""  # Replace this!
client = Groq(api_key=GROQ_API_KEY)

# Prompt to generate input-output pairs
prompt_template = """Generate 10 unique pairs of input text and their corresponding JSON output. Each input should be a natural sentence about a person (e.g., name, age, data pack preference, favorite color, city), and the output should be a JSON object extracting the key attribute from the input. Format each pair as:

Input: <sentence>
Output: <json>

Examples:
Input: My name is Somia
Output: {"name": "Somia"}

Input: I want a 10GB data pack
Output: {"data_pack": "10GB"}

Input: I am 25 years old
Output: {"age": 25}

Now, generate 10 new pairs:
"""

# Function to generate pairs using LLaMA 70B
def generate_input_output_pairs():
    try:
        chat_completion = client.chat.completions.create(
            messages=[{"role": "user", "content": prompt_template}],
            model="llama3-70b-8192",  # LLaMA 3 70B model
            temperature=0.7,  # Balanced creativity and accuracy
            max_tokens=500  # Increased to handle 10 pairs
        )
        response = chat_completion.choices[0].message.content.strip()
        return response
    except Exception as e:
        print(f"Error with Groq API: {e}")
        return ""

# Parse the response into a list of dictionaries
def parse_pairs(response):
    pairs = []
    lines = response.split("\n")
    current_input = None

    for line in lines:
        line = line.strip()
        if line.startswith("Input:"):
            current_input = line.replace("Input:", "").strip()
        elif line.startswith("Output:") and current_input:
            output = line.replace("Output:", "").strip()
            try:
                json.loads(output)  # Validate JSON
                pairs.append({"input": current_input, "output": output})
                current_input = None  # Reset for next pair
            except json.JSONDecodeError:
                print(f"Invalid JSON: {output}")
    return pairs

# Generate synthetic dataset with at least 200 unique rows
synthetic_dataset = []
target_size = 100
attempts = 0
max_attempts = 20  # Prevent infinite loop

while len(synthetic_dataset) < target_size and attempts < max_attempts:
    response = generate_input_output_pairs()
    if response:
        new_pairs = parse_pairs(response)
        synthetic_dataset.extend(new_pairs)
        # Remove duplicates based on input-output tuple
        synthetic_dataset = [dict(t) for t in {tuple(sorted(d.items())) for d in synthetic_dataset}]
        print(f"Attempt {attempts + 1}: Added {len(new_pairs)} pairs. Total unique rows: {len(synthetic_dataset)}")
    attempts += 1
    time.sleep(2)  # Avoid hitting rate limits

# Ensure at least 100 rows (trim if over, warn if under)
if len(synthetic_dataset) >= target_size:
    synthetic_dataset = synthetic_dataset[:target_size]
else:
    print(f"Warning: Only generated {len(synthetic_dataset)} unique rows due to API limits or duplicates.")

# Save as CSV in Colab local storage
df = pd.DataFrame(synthetic_dataset)
save_path = "/content/text_json_2.csv"
df.to_csv(save_path, index=False)
print(f"Dataset saved to {save_path}")
print(f"Generated {len(synthetic_dataset)} rows")

# Preview first 5 rows
print("Sample of the dataset:")
for row in synthetic_dataset[:5]:
    print(f"Input: {row['input']}")
    print(f"Output: {row['output']}")
    print("---")
```
